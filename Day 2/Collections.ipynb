List

```python
numbers = [10, 20, 30, 40, 50, 60, 70]
print("Original List:", numbers)
#Basic Slicing
print("numbers[1:4]   =>", numbers[1:4])     
print("numbers[:3]    =>", numbers[:3])      
print("numbers[3:]    =>", numbers[3:])      
print("numbers[:]     =>", numbers[:])       
#Slicing with Step
print("numbers[::2]   =>", numbers[::2])     
print("numbers[1::2]  =>", numbers[1::2])    
#Negative Indices
print("numbers[-3:]   =>", numbers[-3:])     
print("numbers[:-2]   =>", numbers[:-2])     
#Reverse the List
print("numbers[::-1]  =>", numbers[::-1])    

#list comprehension
squares = [x**2 for x in range(1, 6)]
print("Squares:", squares)

even = [number for number in numbers if number % 2 == 0]
print("Even Numbers:", even)

even = [number for number in numbers if number % 2 != 0]
print("Odd Numbers:", even)

#operations in list 
print("Original list:", numbers)
# Access elements
print("First element:", numbers[0])
print("Last element:", numbers[-1])
# Modify an element
numbers[2] = 35
print("Modifying index 2 (30 to 35):", numbers)
# Add elements
numbers.append(80)
print("After append(80):", numbers)
numbers.insert(1, 15)
print("After insert(15) at index 1:", numbers)
# Remove elements
numbers.remove(35)
print("After remove(35):", numbers)
popped_value = numbers.pop()
print("After pop():", numbers)
print("Popped value:", popped_value)
del numbers[2]
print("After del numbers[2]:", numbers)
# Check existence
print("Is 40 in numbers?", 40 in numbers)
print("Is 99 in numbers?", 99 in numbers)
# Length of list
print("Length of list:", len(numbers))
# Loop through list
print("Looping through list:")
for num in numbers:
    print(num, end=" ")
print()
# Sort and reverse
numbers.sort()
print("Sorted list (ascending):", numbers)
numbers.sort(reverse=True)
print("Sorted list (descending):", numbers)
numbers.reverse()
print("List after reverse():", numbers)
# Copy list
copied_list = numbers.copy()
print("Copied list:", copied_list)
# Concatenate with another list
extra = [100, 110]
combined = numbers + extra
print("Concatenated list:", combined)
# Clear list
numbers.clear()
print("After clear():", numbers)
```

```python
# Creating tuples
mytuple = ("apple", "banana", "cherry")
thistuple = ("apple", "banana", "cherry")
print(thistuple)

# Accessing tuple element by index
thistuple = ("apple", "banana", "cherry")
print(thistuple[1])

# Accessing last element using negative index
thistuple = ("apple", "banana", "cherry")
print(thistuple[-1])

# Slicing the tuple
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:5])  # From index 2 to 4

# Slicing from beginning to index 4 (excluded)
print(thistuple[:4]) 

# Slicing from index 2 to end
print(thistuple[2:])  

# Membership test
thistuple = ("apple", "banana", "cherry")
if "apple" in thistuple:
    print("Yes, 'apple' is in the fruits tuple")

# Modifying tuple by converting to list
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)

# Adding element by tuple concatenation
thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y
print(thistuple)

# Appending item by converting to list
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
print(thistuple)

# Removing item by converting to list
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
print(thistuple)

# Iterating through a tuple
thistuple = ("apple", "banana", "cherry")
for x in thistuple:
    print(x)

# Tuple concatenation
tuple1 = ("a", "b", "c")
tuple2 = (1, 2, 3)
tuple3 = tuple1 + tuple2
print(tuple3)

# Tuple repetition
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2
print(mytuple)
```

Sets

```python
# Creating a set
fruits = {"apple", "banana", "cherry"}
print("Original set:", fruits)
# Adding elements to a set
fruits.add("orange")
print("After adding orange:", fruits)
# Adding multiple elements using update()
fruits.update(["mango", "grape"])
print("After updating with mango and grape:", fruits)

# Removing elements
fruits.remove("banana")  
print("After removing banana:", fruits)

# Discarding elements 
fruits.discard("kiwi")  
print("After discarding kiwi:", fruits)

# Copying a set
copied_fruits = fruits.copy()
print("Copied set:", copied_fruits)

# Clearing all elements
copied_fruits.clear()
print("After clearing copied set:", copied_fruits)

# FROZENSET (immutable set)
frozen = frozenset(["apple", "banana", "cherry"])
print("Frozenset:", frozen)


# UNION, INTERSECTION, DIFFERENCE

setA = {"apple", "banana", "cherry"}
setB = {"cherry", "kiwi", "mango"}

# Union 
union_set = setA.union(setB)
print("Union:", union_set)

# Intersection 
intersection_set = setA.intersection(setB)
print("Intersection:", intersection_set)

# Difference 
difference_set = setA.difference(setB)
print("Difference (A - B):", difference_set)
```

Dictionary 

```python
# Creating a simple dictionary
person = {
    "name": "Hanzla",
    "age": 24,
    "city": "Lahore"
}
# Accessing values
print("Name:", person["name"])
print("Age:", person["age"])
print("City:", person["city"])
# Adding a new key-value pair
person["email"] = "hanzla@gmail.com"
print("After adding email:", person)
# Updating an existing value
person["age"] = 26
print("After updating age:", person)
# Removing a key-value pair
person.pop("city")
print("After removing city:", person)

# NESTED DICTIONARY
students = {
    "student1": {
        "name": "Hanzla",
        "age": 24,
        "grades": [85, 90, 78]
    },
    "student2": {
        "name": "Ahmad",
        "age": 22,
        "grades": [88, 76, 92]
    },
    "student3": {
        "name": "Salman",
        "age": 21,
        "grades": [90, 91, 89]
    }
}

# Accessing nested dictionary values
print("Student1's name:", students["student1"]["name"])
print("Student2's grades:", students["student2"]["grades"])
# Updating a nested value
students["student3"]["age"] = 23
print("Updated student3 age:", students["student3"]["age"])
# Adding a new student
students["student4"] = {
    "name": "Diana",
    "age": 19,
    "grades": [95, 87, 93]
}
print("After adding student4:", students)
# Looping through nested dictionary
print("\n-- Student Details --")
for key, info in students.items():
    print(f"{key}: {info['name']}, Age: {info['age']}, Grades: {info['grades']}")
# pop() vs popitem()
# Using pop() remove item by key
removed_student = students.pop("student2")
print("\nRemoved using pop('student2'):", removed_student)
print("Remaining students:", students)

# Using popitem() - removes last item (LIFO order)
last_item = students.popitem()
print("\nRemoved using popitem():", last_item)
print("Remaining students after popitem():", students)
```
